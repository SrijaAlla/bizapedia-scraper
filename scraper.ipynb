```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
# from selenium import 
```

```python
# opening seperately so can run any cell individually
driver = webdriver.Chrome()
```

```python
url = "https://www.bizapedia.com/"
driver.get(url)

# verify human
time.sleep(5)

jurisdiction_picker = driver.find_element(By.ID, "ctl00_tdJurisdictionPicker")
links = [a.get_attribute('href') for a in jurisdiction_picker.find_elements(By.TAG_NAME, 'a')]

print(links)

```

```python
company_links = []
for link in links:
    driver.get(link)
    time.sleep(2) 

    try:
        trending_table = driver.find_element(By.ID, "ctl00_tdTrendingEntitiesList")
        company_links += [a.get_attribute('href') for a in trending_table.find_elements(By.TAG_NAME, 'a')]

    except Exception as e:
        print(f"Trending table not found on {link}: {e}")

print(len(company_links))
print(company_links)
```

```python
person_links = []
failed_company_links = []
```

```python
print(len(person_links))
print(len(failed_company_links))
```

```python
with open("all_companies.txt", "w") as file:
    for link in company_links:
        file.write(link + "\n")

with open("failed_companies.txt", "w") as file:
    for link in failed_company_links:
        file.write(link + "\n")

with open("person_links.txt", "w") as file:
    for link in person_links:
        file.write(link + "\n")
```

```python
company_links.index("https://www.bizapedia.com/sd/aqreva-llc.html")
```

```python
for company_link in company_links[1081:]:
    print(company_link)
    driver.get(company_link)
    time.sleep(2) 

    try:
        member_tables = driver.find_elements(By.XPATH, "//table[@itemprop='member']")

        for member_table in member_tables:
            # Get the person URL from each member
            person_url = member_table.find_element(By.TAG_NAME, "a").get_attribute("href")
            person_links.append(person_url)
        
    except Exception as e:
        print(f"Could not find member info on {company_link}")
        failed_company_links.append(company_link)

print(len(person_links))
print(person_links)
```

```python
person_info = []
```

```python
for person_link in person_links[582:]:
    driver.get(person_link)
    time.sleep(2)  

    try:
        try :
            # view all link (if exists)
            view_all_data_link = driver.find_element(By.LINK_TEXT, "View All Data")
            driver.execute_script("arguments[0].click();", view_all_data_link)

            time.sleep(4)  
        except: 
            pass

        affiliations = driver.find_elements(By.CLASS_NAME, "groupbox_large")

        count = 1
        for affiliation in affiliations:
            try:
                person_name = affiliation.find_elements(By.XPATH, ".//*[@itemprop='name']")[0].text
            except:
                person_name = ""

            try:
                person_position = affiliation.find_element(By.XPATH, ".//*[@itemprop='jobtitle']").text
            except :
                person_position = ""

            try:
                company_name = affiliation.find_elements(By.XPATH, ".//*[@itemprop='name']")[1].text
            except :
                company_name = ""

            try:
                company_address = affiliation.find_element(By.XPATH, ".//*[@itemprop='address']").text
            except :
                company_address = ""


            # Print the details
            print(f"Count: {count}")
            print(f"Person Name: {person_name}")
            print(f"Position: {person_position}")
            print(f"Company Name: {company_name}")
            print(f"Address: {company_address}")
            print("-----")
            count += 1

            # store details
            person_data = {
                "Person Name": person_name,
                "Position": person_position,
                "Company Name": company_name,
                "Address": company_address
            }
            person_info.append(person_data)

    except Exception as e:
        print(f"Could not retrieve information from {person_link}: {e}")
```
